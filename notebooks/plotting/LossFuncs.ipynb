```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
def huber(x):
    if np.abs(x) < 1:
        return x**2
    else:
        return np.abs(x)
```

```python
x = np.linspace(-1.5,1.5,1000)
```

```python
import sys
sys.path.append('../../')

import denn.utils
```

```python
plt.figure(figsize=(6,4))
plt.rc('axes', titlesize=22, labelsize=22)
plt.rc('legend', fontsize=18)
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)
plt.rcParams['text.usetex'] = True
plt.plot(x, x**2, color='blue', linestyle='dashed', lw=2.5, label='$L_2$', alpha=0.7)
plt.plot(x, np.abs(x), color='limegreen', linestyle='dashdot', lw=2.5, label="$L_1$", alpha=.7)
plt.plot(x, [huber(xi) for xi in x], color='crimson', label='Huber', lw=2.5, alpha=0.7)
plt.xlabel("$e$")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig("huber_l2_l1.png", bbox_inches='tight', dpi=300);
```
