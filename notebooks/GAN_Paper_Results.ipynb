# Varying obs_every

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('axes', titlesize=15, labelsize=12)
plt.rc('legend', fontsize=12)
```

```python
lag = pd.read_csv('~/Downloads/LAG_obsevery_mse_results.csv', index_col=0)
```

```python
gan = pd.read_csv('/Users/dylanrandle/Downloads/GAN_obsevery_mse_results.csv', index_col=0)
```

```python
lag['mses'] = lag.mses.apply(lambda x: np.array([float(i) for i in x.strip('][').split(', ')]))
gan['mses'] = gan.mses.apply(lambda x: np.array([float(i) for i in x.strip('][').split(', ')]))
```

```python
lag = lag.sort_values(by='obs_every')
gan = gan.sort_values(by='obs_every')
```

```python
plt.figure(figsize=(10,6))
plt.errorbar(lag.obs_every, [np.mean(mse[:3]) for mse in lag.mses], [np.std(mse[:3]) for mse in lag.mses], label='MSE')
plt.errorbar(gan.obs_every, [np.mean(mse) for mse in gan.mses], [np.std(mse) for mse in gan.mses], label='GAN')
plt.yscale('log')
plt.xlabel('Observer Gap')
plt.ylabel('Mean Square Error')
plt.legend()
```
