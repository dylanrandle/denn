# RK4

```python
import sys
sys.path.append('../../')

from denn.rk4 import rk4
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma, A = 0.1, 0.1

means = [[0.74507886, 0.3602802 ],
            [0.40147605, 0.06139579],
            [0.94162198, 0.46722697],
            [0.79110703, 0.8973808 ],
            [0.64732527, 0.07095655],
            [0.10083943, 0.31935057],
            [0.24929806, 0.60499613],
            [0.11377013, 0.42598647],
            [0.85163671, 0.26495608],
            [0.18439795, 0.31438099]]

def rays_deriv(t, params):
    x, y, px, py = params
    Vx, Vy = 0, 0
    for mean in means:
        Vx += A*np.exp(-(((x-mean[0])**2 + (y-mean[1])**2) / sigma**2)/2) * (x-mean[0])/sigma**2 
        Vy += A*np.exp(-(((x-mean[0])**2 + (y-mean[1])**2) / sigma**2)/2) * (y-mean[1])/sigma**2 
    rhs1 = px
    rhs2 = py
    rhs3 = -Vx
    rhs4 = -Vy
    return np.array([rhs1, rhs2, rhs3, rhs4])

init_cond = [0, 0.3, 1, 0]
```

```python
t, sol = rk4(rays_deriv, [0, 1], init_cond, 400)
```

```python
from denn.ode_problems import RaysEquations
rays_sol = RaysEquations(t_min=0, t_max=1, n=400).get_solution(t)
```

```python
sol.shape, rays_sol.shape
```

```python
mse = np.mean( (sol - rays_sol.numpy())**2 )
```

```python
mse
```

```python
fig, ax = plt.subplots(1, 4, figsize=(15,4))
ax = ax.ravel()

for i in range(4):
    ax[i].plot(t, sol[:, i])
    ax[i].plot(t, rays_sol[:, i], '--')
    
plt.show()
```

# Euler Method to compare

```python
init_cond = [sir.S0, sir.I0, sir.R0]
t = sir.grid.detach().numpy().flatten()
```

```python
def euler( x0, y, h, x, func ): 
    temp = -0
  
    # Iterating till the point at which we 
    # need approximation 
    while x0 < x: 
        temp = y 
        y = y + h * func(x0, y) 
        x0 = x0 + h 
        
    return y
```

```python
h = float((t[1] - t[0]))
h
```

```python
yeul = []
init_cond = [0.7, 0.3, 0.0]
for p in t:
    yeu = euler(0, init_cond, h, p, sir._sir_system) 
    yeul.append(yeu)
yeul = np.vstack(yeul)
```

```python
plt.plot(t, yeul[:,0])
plt.plot(t, yeul[:,1])
plt.plot(t, yeul[:,2]);
```

```python
plt.figure(figsize=(10,6))
t = np.linspace(0, 10, 200)
alpha=0.8
plt.plot(t, snum+inum+rnum, label='RK4', color='black', linestyle='dotted', alpha=alpha)
plt.plot(t, spred+ipred+rpred, label='GAN', color='crimson', alpha=alpha)
plt.plot(t, np.sum(yeul, axis=1), label='euler', color='orange', alpha=alpha)
plt.plot(t, sl2+il2+rl2, label='$L_2$', color='violet', linestyle='solid', alpha=alpha)
plt.plot(t, sl1+il1+rl1, label='$L_1$', color='limegreen', linestyle='dashdot', alpha=alpha)
plt.plot(t, shu+ihu+rhu, label='Huber', color='blue', linestyle='dashed', alpha=alpha)
plt.xlabel('Time ($t$)')
plt.ylabel('Population ($N$)')
plt.legend(loc='lower right')
plt.grid()
plt.ylim([0.985, 1.005]);
```
